```python
class SalaryNotInRangeError(Exception):
    """Exception raised for errors in the input salary.

        message -- explanation of the error
    """

    def __init__(self,  message="Salary is not in range"):
        self.message = message
        super().__init__(self.message)

def check_salary(salary):
    try:
        if not 10000 < salary < 100000:
            raise SalaryNotInRangeError
        else:
            return "Congratulations!!"
    
    except SalaryNotInRangeError as e:
        return str(e)
```

```python
class SalaryNotInRangeError(Exception):
    """Exception raised for errors in the input salary.

        message -- explanation of the error
    """

    def __init__(self, salary, message="Salary is not in range"):
        self.salary = salary
        self.message = message
        super().__init__(self.message)

def check_salary(salary):
  try:
    if (salary in range(1000,100000)):
        raise SalaryNotInRangeError(salary)
    else:
      return "Congratulations!!"
    
  except SalaryNotInRangeError as e:
    return str(e)
```
