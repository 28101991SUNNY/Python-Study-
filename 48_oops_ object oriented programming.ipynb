```python
class Account:
    
    counter = 34900
    
    def __init__(self, opening_balance = 0):
        
        self.balance = opening_balance
        
        Account.counter += 1
        self.Account_Number = self.counter
        
        self.num_of_transaction = 0
        self.max_transaction = 5
        
        
    def __str__(self):
        return f"Acc Number:<{self.Account_Number}> has Balance:{self.balance}" 
    
    def deposit(self,amount):
        if self.num_of_transaction < self.max_transaction:
            if amount > 0:
                self.balance += amount
                self.num_of_transaction += 1
        else:
            return f"Max Transaction Limit Achieved and balance is {self.balance}"
        
        
    def withdraw(self,amount):
        if self.num_of_transaction < self.max_transaction:  
            if self.balance >= amount and amount > 0:
                self.balance -= amount
                self.num_of_transaction += 1
        else:
            return f"Max Transaction Limit Achieved and balance is {self.balance}"
        
    def __repr__(self):
        return f"{self.Account_Number}"
        
```

```python
class SavingsAccount(Account):
    pass
class CurrentAccount(Account):    # inheritated from parent class: Account 
    pass
```

```python
s1 = SavingsAccount()
c1 = CurrentAccount()
print(s1)
print(c1)

# c1 and s1 are the instances of currentaccount and savingaccount class 
# respectively
# and they are inheriting the properties from praent class called Account
```

```python
class Account:
    
    counter = 0
    
    def __init__(self, opening_balance = 0):
        
        self.__balance = opening_balance
        
        Account.counter += 1
        self.Account_Number = Account.counter
        
        self.num_of_transaction = 0
        self.max_transaction = 5
        
    
    
    def deposit(self,amount):
        if self.num_of_transaction < self.max_transaction:
            if amount > 0:
                self.__balance += amount
                self.num_of_transaction += 1
        else:
            return f"Max Transaction Limit Achieved and balance is {self.__balance}"
        
        
    def withdraw(self,amount):
        if self.num_of_transaction < self.max_transaction:  
            if self.__balance >= amount and amount > 0:
                self.__balance -= amount
                self.num_of_transaction += 1
        else:
            return f"Max Transaction Limit Achieved and balance is {self.__balance}"
        
    def __repr__(self):
        return f"{self.Account_Number}"
        
    def __str__(self):
        return f"Acc Number:<{self.Account_Number}> has Balance:{self.__balance}" 
    
    def getInterest(self):
        pass
        

        
class SavingsAccount(Account):
    interest_rate_savings = 0.10
    def __init__(self):
        super().__init__()
    
    def getInterest(self):
        interest = self._Account__balance * SavingsAccount.interest_rate_savings
        print(f"Interest on Account {self.Account_Number} is {interest}")

class CurrentAccount(Account):
    interest_rate_current = 0.03
    def __init__(self):
        
        super().__init__()  # super() reference to parent class which gives access to all child classes
       # print(super())
        self.max_transaction = 10
        
    def getInterest(self):
        interest = self._Account__balance *  CurrentAccount.interest_rate_current
        print(f"Interest on Account {self.Account_Number} is {interest}")
```

```python
class Props:
    def __init__(self, x, y):
        self.x = x
        self.y = y
class Ball(Props):
    def __init__(self, x, y):
        super().__init__(x, y)
        self.shape = "o"
        self.number = 1
        self.vely = 0
        self.velx = 0

    def get_x(self):
        return self.x
 
    def get_y(self):
        return self.y

    def get_velx(self):
        return self.velx
 
    def get_vely(self):
        return self.vely
    
    def inc_velx(self):
        self.velx = self.velx+2
    
    def inc_vely(self):
        self.vely = self.vely+1
```

```python
obj_ball = Ball(27,3)

bx = obj_ball.get_x()
by = obj_ball.get_y()
b_velx = obj_ball.get_velx()
b_vely = obj_ball.get_vely()

print(bx, by, b_velx, b_vely)
```

```python
class Circle:
    pi = 3.14

    def __init__(self, diameter):
        self.radius = diameter / 2

    def circumference(self):
        return 2 * self.pi * self.radius
```
